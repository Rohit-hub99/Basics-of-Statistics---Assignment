# 1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.
  - Types of Data in Statistics
     - Data is classified into two main types: qualitative and quantitative data.

  1. Qualitative (Categorical) Data
    - Qualitative data describes qualities, attributes, or categories that cannot be measured numerically but can be classified. It is divided into nominal and ordinal scales.

    - Nominal Scale (No order, just labels)
      - Data that consists of names, labels, or categories without any meaningful order.

      Examples:

       - Eye color (blue, green, brown)
       - Gender (male, female, other)
       - Types of cars (sedan, SUV, hatchback)

    - Ordinal Scale (Ordered, but differences are not meaningful)
      - Data that can be arranged in a meaningful order, but the differences between values are not measurable or equal.

      Examples:

       - Education level (High school, Bachelor's, Master's, Ph.D.)
       - Customer satisfaction ratings (poor, average, good, excellent)
       - Military ranks (Private, Sergeant, Captain, General)

  2. Quantitative (Numerical) Data:
    - Quantitative data consists of numbers and measurable values. It is further divided into interval and ratio scales.

    - Interval Scale (Ordered with meaningful differences, but no true zero)
      - Data has equal intervals between values, but zero does not mean the absence of the property.

      Examples:

      - Temperature in Celsius or Fahrenheit (0¬∞C does not mean "no temperature")
      - Years on a timeline (2000 AD, 2020 AD, etc.)
      - IQ scores

    - Ratio Scale (Ordered with meaningful differences and a true zero)
      - Data has all the properties of an interval scale, but zero represents the absolute absence of a quantity.

      Examples:

      - Height and weight (0 cm means no height, 0 kg means no weight)
      - Age (0 years means newborn)
      - Income (0 income means no money)


# 2.  What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.
  - Measures of Central Tendency
    - Measures of central tendency summarize a dataset by identifying a central value. The three main measures are Mean, Median, and Mode.

   1. Mean (Arithmetic Average)
    - The mean is the sum of all values divided by the total number of values.

     - Example:
      - Suppose a class has test scores: 70, 80, 85, 90, 95
       - To calculate mean first we will add all the scores like 70+80+85+90+95 / total number of subjects(5) = 420/5 = 84

     - When to Use the Mean:
       - Best for continuous and normally distributed data.
       - Used when no extreme values (outliers) are present.
       - Common in scientific research, economics, and business (e.g., average income, GDP, stock prices)

     - When NOT to Use the Mean:
       - If the dataset has outliers, they can skew the mean.
       - Example: Income data (if a billionaire is included, the mean income may not represent the majority).
    
    2. Median (Middle Value)
     - The median is the middle value when data is arranged in ascending order. If there is an even number of values, the median is the average of the two middle numbers.

      - Example:
        - Scores: 60, 75, 80, 85, 90
        - Median = 80 (middle value)

        - For an even dataset: 50, 55, 60, 70, 75, 80
        = median = 60+70/2 = 65

      - When to Use the Median:
       - Best when data is skewed or contains outliers.
       - Used in income distribution, house prices, and salaries (e.g., median salary is better than mean salary in analyzing wages).

      - When NOT to Use the Median:
       - Not ideal for small datasets with evenly spaced values.

    3. Mode (Most Frequent Value)
     - The mode is the most frequently occurring value in a dataset. There can be one mode (unimodal), two modes (bimodal), or more (multimodal).

       - Example:
        - Unimodal: Scores = 60, 70, 75, 80, 80, 85 ‚Üí Mode = 80
        - Bimodal: Scores = 10, 20, 20, 30, 40, 40, 50 ‚Üí Modes = 20 and 40
        - No Mode: Scores = 5, 10, 15, 20 (Each value appears once)

      - When to Use the Mode:
       - Best for categorical or nominal data (e.g., most common shoe size, most preferred coffee flavor).
       - Used in marketing, fashion, and voting trends (e.g., most sold smartphone brand).

      - When NOT to Use the Mode:
       - If all values appear equally, mode is not useful.
       - Not useful for continuous data like height or weight.

# 4.  What is a box plot, and what can it tell you about the distribution of data?
  - Box Plot (Box-and-Whisker Plot)
    - A box plot is a graphical representation of data distribution that shows the spread, central tendency, and presence of outliers. It is useful for comparing distributions and identifying skewness.

   - Key Components of a Box Plot
     - Median (Q2) ‚Üí The middle value of the dataset.
     - First Quartile (Q1) ‚Üí The median of the lower half (25th percentile).
     - Third Quartile (Q3) ‚Üí The median of the upper half (75th percentile).
     - Interquartile Range (IQR) ‚Üí The range between Q1 and Q3:
        - IQR=Q3‚àíQ1
     - Whiskers ‚Üí Extend to the minimum and maximum values within 1.5 √ó IQR from Q1 and Q3.
     - Outliers ‚Üí Data points beyond 1.5 √ó IQR, plotted as individual dots.


# 5.  Discuss the role of random sampling in making inferences about populations
  - Role of Random Sampling in Making Inferences About Populations
   - Random sampling is a technique used in statistics where a subset of a population is selected in such a way that every individual has an equal chance of being chosen. It helps ensure that the sample represents the entire population, making statistical inferences reliable.

   - How Random Sampling Supports Inference
     - Estimation of Population Parameters
       - Example: If a company surveys 200 employees out of 5,000, the mean salary of the sample can estimate the mean salary of the entire workforce.
     - Hypothesis Testing
       - Example: A researcher tests whether a new drug is effective by randomly selecting patients. If the results show significant improvement, the inference is that the drug works for the larger population.
     - Confidence Intervals
       - A random sample allows us to construct confidence intervals, which provide a range where the true population parameter likely falls.
       - Example: If a survey finds that 60% ¬± 3% of people prefer a brand, we are 95% confident that between 57% and 63% of the population prefers it.
     - Reducing Sampling Error
       - A larger random sample reduces sampling error and increases accuracy.



# 6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
  - Skewness measures the asymmetry of a dataset‚Äôs distribution. It tells us whether data is symmetrically distributed or leans towards one side of the mean.

  - Types of Skewness
   - 1. Symmetrical Distribution (Zero Skewness)
   - 2. Positive Skewness (Right-Skewed)
   - 3. Negative Skewness (Left-Skewed)
   
  - How Skewness Affects Data Interpretation
    - Impact on Measures of Central Tendency
     - In right-skewed data, the mean is higher than the median, so the median is a better measure of central tendency.
     - In left-skewed data, the mean is lower than the median, so the median should be used.
    - Effect on Decision-Making
     - Right-skewed data (e.g., income) means most people earn less than the mean, so policies should focus on lower-income groups.
     - Left-skewed data (e.g., age at retirement) suggests most people retire early, affecting pension planning.
    - Influence on Statistical Analysis
     - Many statistical tests assume normality (zero skewness).
     - If data is skewed, we may need log transformation or non-parametric tests.



# 7. What is the interquartile range (IQR), and how is it used to detect outliers?
  - Interquartile Range (IQR)
   - The Interquartile Range (IQR) is a measure of statistical dispersion that describes the spread of the middle 50% of a dataset. It is the difference between the third quartile (Q3) and the first quartile (Q1):
         - IQR=Q3‚àíQ1
   - Q1 (First Quartile, 25th percentile) ‚Üí The median of the lower half of the dataset.
   - Q3 (Third Quartile, 75th percentile) ‚Üí The median of the upper half of the dataset.
   - IQR represents the range where the central 50% of values lie.

  - Using IQR to Detect Outliers
   - Outliers are data points that fall significantly outside the normal range. The 1.5 √ó IQR Rule helps identify these outliers:
     
     - Outlier Boundaries
      - Lower Bound:
            - Q1‚àí(1.5√óIQR)
      - Upper Bound:
            - Q3+(1.5√óIQR)
      - Any data point below the lower bound or above the upper bound is considered an outlier.
   
   

# 8.  Discuss the conditions under which the binomial distribution is used
  - The binomial distribution is used when a random experiment meets specific conditions. It models the number of successes in a fixed number of independent trials, each with the same probability of success.

   - Conditions for Using Binomial Distribution
    - 1. Fixed Number of Trials (

n)
     - The experiment is repeated a specific number of times.
     - Example: Tossing a coin 10 times (trials = 10).
   
    - 2. Only Two Possible Outcomes (Success or Failure)
     - Each trial has exactly two outcomes: Success (S) or Failure (F).
     - Example:
       - Success = "Heads", Failure = "Tails" in a coin toss.
       - Success = "Pass", Failure = "Fail" in an exam.
   
    - 3. Constant Probability of Success (
ùëù
p)
     - The probability of success remains the same for each trial.
     - Example: If a die roll is considered successful when a "6" appears, then p = 1/6 for all rolls.

    - 4. Independent Trials
     - The outcome of one trial does not affect another.
     - Example: The result of one coin toss does not impact the next.



# 9.  Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
  - A normal distribution, also called a Gaussian distribution, is a bell-shaped probability distribution that is symmetric around the mean. It is widely used in statistics because many natural phenomena follow this distribution, such as heights, IQ scores, and exam results.
  
  - Properties of Normal Distribution
    - Bell-Shaped & Symmetric
     - The curve is perfectly symmetric around the mean (Œº).
     - Mean = Median = Mode in a normal distribution.

    - Defined by Mean (Œº) & Standard Deviation (œÉ)
     - The mean (Œº) determines the center of the curve.
     - The standard deviation (œÉ) determines the spread of the curve.

    - Total Probability = 1
     - The total area under the curve is 1 (or 100%).

    - Follows the Empirical Rule (68-95-99.7 Rule)
     - Most data falls within three standard deviations from the mean.

    - Asymptotic Nature
     - The tails of the curve never touch the x-axis; they extend infinitely.

  
  - Empirical Rule (68-95-99.7 Rule)
    - The empirical rule describes how data is distributed in a normal distribution:
     - 68% of data falls within 1 standard deviation (Œº ¬± 1œÉ).
     - 95% of data falls within 2 standard deviations (Œº ¬± 2œÉ).
     - 99.7% of data falls within 3 standard deviations (Œº ¬± 3œÉ).
   
    - Visual Representation of the Empirical Rule:
      -  If Œº = 100 and œÉ = 15 (e.g., IQ scores):
       - 68% of people have IQs between 85 and 115.
       - 95% of people have IQs between 70 and 130.
       - 99.7% of people have IQs between 55 and 145.

```python
# 12.  Provide an example dataset, calculate both covariance and correlation, and interpret the results.

import pandas as pd
import numpy as np

# Create the dataset as a dictionary
data = {
    "Student": [1, 2, 3, 4, 5],
    "Hours_Studied (X)": [2, 3, 5, 7, 9],
    "Exam_Score (Y)": [50, 55, 65, 70, 85]
}

# Convert to a pandas DataFrame
df = pd.DataFrame(data)

# Display the dataset
print(df)
```

```python
# Compute covariance matrix
cov_matrix = df[["Hours_Studied (X)", "Exam_Score (Y)"]].cov()

# Extract covariance value
covariance = cov_matrix.iloc[0, 1]
print(f"Covariance: {covariance:.2f}")
```

```python
# Compute correlation matrix
correlation_matrix = df[["Hours_Studied (X)", "Exam_Score (Y)"]].corr()

# Extract correlation value
correlation = correlation_matrix.iloc[0, 1]
print(f"Correlation: {correlation:.2f}")
```
